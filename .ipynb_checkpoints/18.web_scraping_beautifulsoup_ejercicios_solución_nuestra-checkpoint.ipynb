```python
url = "https://www.recetasgratis.net/Recetas-de-Ensaladas-listado_receta-4_1.html"

response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
soup.find("div", class_= "resultado link").find("a")["href"]
```

```python
urls = list()

for bs in soup.find_all("div", class_= "resultado link"):
    url = bs.find("a")["href"]
    
    urls.append(url)
    
urls
```

```python
response_receta = requests.get(urls[0])

soup_receta = BeautifulSoup(response_receta.text, "html.parser")

soup_receta
```

```python
#Instrucciones
instrucciones = list()

for bs in soup_receta.find_all("div", class_ = "apartado"):
    instruccion = bs.find("p")
    try:
        orden = bs.find("div", class_ = "orden").text
    except:
        orden = "*"
    
    instrucciones.append(orden + ". "+ instruccion.text) #.text quita etiquetas html
    
instrucciones
```

```python
pasos = list()

for paso in instrucciones:
    if not(paso.startswith("*")):
        pasos.append(paso)

"\n".join(pasos)
```

```python
urls = list()
nombres = list()
descripciones = list()
comensales = list()
tiempos = list()
categorias = list()
dificultades = list()
ingredientes = list()
instrucciones = list()

url = "https://www.recetasgratis.net/Recetas-de-Ensaladas-listado_receta-4_1.html"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")



for bs in soup.find_all("div", class_= "resultado link"):
    url = bs.find("a")["href"]
    
    urls.append(url)
    

    
for url in urls:
    print(url) #Nos aseguramos de que entramos en todas las páginas
    
    response = requests.get(url)
    soup_receta = BeautifulSoup(response.text, "html.parser")

    #Nombre
    try:
        nombre = soup_receta.find("h1").text
    except:
        nombre = np.nan
        
        
    #Descripcion
    try:
        parrafos = list()

        for bs in soup_receta.find("div", class_ = "intro").find_all("p"):
            parrafo = bs.text

            parrafos.append(parrafo)

        descripcion = "\n".join(parrafos)
    except:
        descripcion = np.nan
        
    #Comensales
    try:
        comensal = int(soup_receta.find("div", class_ = "properties").find("span", class_ = "property comensales").text.split(" ")[0])
    except:
        comensal = np.nan
        
    
    #Tiempo
    try:
        tiempo = soup_receta.find("div", class_ = "properties").find("span", class_ = "property duracion").text
    except:
        tiempo = np.nan
        
    #Categoria
    try:
        categoria = soup_receta.find("div", class_ = "properties").find("span", class_ = "property para").text
    except:
        categoria = np.nan
        
    #Dificultad
    try:
        dificultad = soup_receta.find("div", class_ = "properties").find("span", class_ = "property dificultad").text
    except:
        dificultad = np.nan
        
    #Ingredientes
    try:
        ingreds = list()

        for bs in soup_receta.find("div", class_ = "ingredientes").find_all("li", class_ = "ingrediente"):
            ingred = bs.text.strip()

            ingreds.append(ingred)

        ingrediente = ", ".join(ingreds)
    except:
        ingrediente = np.nan
        
    #Instrucciones
    try:
        instrs = list()

        for bs in soup_receta.find_all("div", class_ = "apartado"):
            instr = bs.find("p")
            try:
                orden = bs.find("div", class_ = "orden").text
            except:
                orden = "*"

            instrs.append(orden + ". "+ instr.text) #.text quita etiquetas html


            pasos = list()

        for paso in instrs:
            if not(paso.startswith("*")):
                pasos.append(paso)

        instruccion = "\n".join(pasos)
    except:
        instruccion = np.nan
        
        
    nombres.append(nombre)
    descripciones.append(descripcion)
    comensales.append(comensal)
    tiempos.append(tiempo)
    categorias.append(categoria)
    dificultades.append(dificultad)
    ingredientes.append(ingrediente)
    instrucciones.append(instruccion)

    sleep(1)
```

- Repetir el ejercicio para todas las recetas de todas las páginas.

```python
paginas = list()

for i in range(1,100):
    url = f"https://www.recetasgratis.net/Recetas-de-Ensaladas-listado_receta-4_{i}.html"
    response = requests.get(url)
    if (i != 1) and (response.url == "https://www.recetasgratis.net/Recetas-de-Ensaladas-listado_receta-4_1.html"):
        break
    
    paginas.append(url)
    print(f"https://www.recetasgratis.net/Recetas-de-Ensaladas-listado_receta-4_{i}.html")
    
    sleep(1)
```

```python
todos_urls = list()

for pagina in paginas:
    
    response = requests.get(pagina)
    
    soup = BeautifulSoup(response.text, "html.parser")

    for bs in soup.find_all("div", class_= "resultado link"):
        url = bs.find("a")["href"]

        todos_urls.append(url)
        
    sleep(1)

```

```python
len(todos_urls)
```

```python
nombres = list()
descripciones = list()
comensales = list()
tiempos = list()
categorias = list()
dificultades = list()
ingredientes = list()
instrucciones = list()
      


for url in todos_urls:
    print(url) #Nos aseguramos de que entramos en todas las páginas

    response = requests.get(url)
    soup_receta = BeautifulSoup(response.text, "html.parser")

    #Nombre
    try:
        nombre = soup_receta.find("h1").text
    except:
        nombre = np.nan


    #Descripcion
    try:
        parrafos = list()

        for bs in soup_receta.find("div", class_ = "intro").find_all("p"):
            parrafo = bs.text

            parrafos.append(parrafo)

        descripcion = "\n".join(parrafos)
    except:
        descripcion = np.nan

    #Comensales
    try:
        comensal = int(soup_receta.find("div", class_ = "properties").find("span", class_ = "property comensales").text.split(" ")[0])
    except:
        comensal = np.nan


    #Tiempo
    try:
        tiempo = soup_receta.find("div", class_ = "properties").find("span", class_ = "property duracion").text
    except:
        tiempo = np.nan

    #Categoria
    try:
        categoria = soup_receta.find("div", class_ = "properties").find("span", class_ = "property para").text
    except:
        categoria = np.nan

    #Dificultad
    try:
        dificultad = soup_receta.find("div", class_ = "properties").find("span", class_ = "property dificultad").text
    except:
        dificultad = np.nan

    #Ingredientes
    try:
        ingreds = list()

        for bs in soup_receta.find("div", class_ = "ingredientes").find_all("li", class_ = "ingrediente"):
            ingred = bs.text.strip()

            ingreds.append(ingred)

        ingrediente = ", ".join(ingreds)
    except:
        ingrediente = np.nan

    #Instrucciones
    try:
        instrs = list()

        for bs in soup_receta.find_all("div", class_ = "apartado"):
            instr = bs.find("p")
            try:
                orden = bs.find("div", class_ = "orden").text
            except:
                orden = "*"

            instrs.append(orden + ". "+ instr.text) #.text quita etiquetas html


            pasos = list()

        for paso in instrs:
            if not(paso.startswith("*")):
                pasos.append(paso)

        instruccion = "\n".join(pasos)
    except:
        instruccion = np.nan


    nombres.append(nombre)
    descripciones.append(descripcion)
    comensales.append(comensal)
    tiempos.append(tiempo)
    categorias.append(categoria)
    dificultades.append(dificultad)
    ingredientes.append(ingrediente)
    instrucciones.append(instruccion)

    sleep(0.5)
```

```python
df = pd.DataFrame()

df['nombres'] = nombres
df['descripciones'] = descripciones
df['comensales'] = comensales
df['tiempos'] = tiempos
df['categorias'] = categorias
df['dificultades'] = dificultades
df['ingredientes'] = ingredientes
df['instrucciones'] = instrucciones
df['urls'] = todos_urls

df
```

```python
df.to_csv('ejercicio_recetas_ensaladas.csv', index = False, sep = ',')
```
